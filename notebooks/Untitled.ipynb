```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/contributions-AIConsult2020.csv', encoding='latin-1', sep=';', on_bad_lines='skip')
print(data.shape)
```

```python
data['User type'].value_counts().plot(kind='barh', figsize=(8,5), title='Responses by Actor Type')
plt.tight_layout()
plt.show()
```

```python
pdf_col = 'You can upload a document here:\n\n'
pdf_by_actor = data.groupby('User type')[pdf_col].apply(lambda x: x.notna().sum()).sort_values(ascending=False)
pdf_by_actor.plot(kind='barh', figsize=(8,5), title='PDF Submissions by Actor Type')
plt.tight_layout()
plt.show()
```
